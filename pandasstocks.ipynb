### Loading Data and make first row column headers


```python
import pandas as pd
import numpy as np

stocks = pd.read_csv('stocks.csv', sep=r'\s*,\s*', header=0, encoding='ascii', engine='python')

```

### Add New columns with the difference in 30 minute intervals starting at noon

```python
stocks['Diff12-12:30'] = stocks['12:30'] - stocks['noon']
stocks['Diff12:30-1'] = stocks['1'] - stocks['12:30']
stocks['Diff1-1:30'] = stocks['1:30'] - stocks['1']
stocks['Diff1:30-2'] = stocks['2'] - stocks['1:30']
stocks['Diff2-2:30'] = stocks['2:30'] - stocks['2']
stocks['Diff2:30-3'] = stocks['3'] - stocks['2:30']
stocks['Diff3-3:30'] = stocks['3:30'] - stocks['3']
stocks['Diff3:30-4'] = stocks['4'] - stocks['3:30']
stocks['Diff12-4'] = stocks['4'] - stocks['noon']
stocks1 = stocks[['ticker','date','Diff12-12:30','Diff12:30-1','Diff1-1:30','Diff1:30-2','Diff2-2:30', 'Diff2:30-3', 'Diff3-3:30', 'Diff3:30-4', 'Diff12-4']]
print(stocks1)
```

### Group by tickers with the most occurrences

```python
stocks2 = stocks1.groupby('ticker').size().sort_values(ascending = False).reset_index(name='ocurrences')
print (stocks2)



```

### Get grouped by count where difference is >0.5 by ticker

```python
## group by and count occurences of a difference of 0.5 or more for each ticker
                                                
stocks3 = stocks1[stocks1['Diff12-12:30'] > 0.5].groupby('ticker', as_index = False)['Diff12-12:30'].count()
stocks6 = stocks1[stocks1['Diff12:30-1'] > 0.5].groupby('ticker', as_index = False)['Diff12:30-1'].count()
stocks7 = stocks1[stocks1['Diff1-1:30'] > 0.5].groupby('ticker', as_index = False)['Diff1-1:30'].count()
stocks8 = stocks1[stocks1['Diff1:30-2'] > 0.5].groupby('ticker', as_index = False)['Diff1:30-2'].count()
stocks9 = stocks1[stocks1['Diff2-2:30'] > 0.5].groupby('ticker', as_index = False)['Diff2-2:30'].count()
stocks10 = stocks1[stocks1['Diff2:30-3'] > 0.5].groupby('ticker', as_index = False)['Diff2:30-3'].count()
stocks11 = stocks1[stocks1['Diff3-3:30'] > 0.5].groupby('ticker', as_index = False)['Diff3-3:30'].count()
stocks12 = stocks1[stocks1['Diff3:30-4'] > 0.5].groupby('ticker', as_index = False)['Diff3:30-4'].count()
stocks13 = stocks1[stocks1['Diff12-4'] > 0.5].groupby('ticker', as_index = False)['Diff12-4'].count()

## Sort these by the count
stocks4 = stocks3.sort_values(by=['Diff12-12:30'], ascending=False)
stocks66 = stocks6.sort_values(by=['Diff12:30-1'], ascending=False)
stocks77 = stocks7.sort_values(by=['Diff1-1:30'], ascending=False)
stocks88 = stocks8.sort_values(by=['Diff1:30-2'], ascending=False)
stocks99 = stocks9.sort_values(by=['Diff2-2:30'], ascending=False)
stocks1010 = stocks10.sort_values(by=['Diff2:30-3'], ascending=False)
stocks1111 = stocks11.sort_values(by=['Diff3-3:30'], ascending=False)
stocks1212 = stocks12.sort_values(by=['Diff3:30-4'], ascending=False)
stocks1313 = stocks13.sort_values(by=['Diff12-4'], ascending=False)
```

### Merge Total Count Dataset with Total Count where Diff >0.5 Dataset

```python
stocks5 = pd.merge(stocks4, stocks2, on="ticker")
stocks5['patternpercentage'] = stocks5['Diff12-12:30']/stocks5['ocurrences']
stocks5sorted = stocks5.sort_values(by=['patternpercentage'], ascending=False)

stocks666 = pd.merge(stocks66, stocks2, on="ticker")
stocks666['patternpercentage'] = stocks666['Diff12:30-1']/stocks666['ocurrences']
stocks666sorted = stocks666.sort_values(by=['patternpercentage'], ascending=False)

stocks777 = pd.merge(stocks77, stocks2, on="ticker")
stocks777['patternpercentage'] = stocks777['Diff1-1:30']/stocks777['ocurrences']
stocks777sorted = stocks777.sort_values(by=['patternpercentage'], ascending=False)

stocks888 = pd.merge(stocks88, stocks2, on="ticker")
stocks888['patternpercentage'] = stocks888['Diff1:30-2']/stocks888['ocurrences']
stocks888sorted = stocks888.sort_values(by=['patternpercentage'], ascending=False)

stocks999 = pd.merge(stocks99, stocks2, on="ticker")
stocks999['patternpercentage'] = stocks999['Diff2-2:30']/stocks999['ocurrences']
stocks999sorted = stocks999.sort_values(by=['patternpercentage'], ascending=False)

stocks101010 = pd.merge(stocks1010, stocks2, on="ticker")
stocks101010['patternpercentage'] = stocks101010['Diff2:30-3']/stocks101010['ocurrences']
stocks101010sorted = stocks101010.sort_values(by=['patternpercentage'], ascending=False)

stocks111111 = pd.merge(stocks1111, stocks2, on="ticker")
stocks111111['patternpercentage'] = stocks111111['Diff3-3:30']/stocks111111['ocurrences']
stocks111111sorted = stocks111111.sort_values(by=['patternpercentage'], ascending=False)

stocks121212 = pd.merge(stocks1212, stocks2, on="ticker")
stocks121212['patternpercentage'] = stocks121212['Diff3:30-4']/stocks121212['ocurrences']
stocks121212sorted = stocks121212.sort_values(by=['patternpercentage'], ascending=False)

stocks131313 = pd.merge(stocks1313, stocks2, on="ticker")
stocks131313['patternpercentage'] = stocks131313['Diff12-4']/stocks131313['ocurrences']
stocks131313sorted = stocks131313.sort_values(by=['patternpercentage'], ascending=False)
```
